### Effective Programming

a) What is a drawback of the top down approach?

One drawback of the top-down approach is that projects can be delayed or spiral out of control if resources/time-spent for plans are estimated incorrectly. Say you have to design a system, and you know the high-level specifications of it. If the time it takes to implement a core, dependent lower-level piece of the project is underestimated, it could cause a bottleneck in the project timeline.
Additionally, changes in project specifications mid-development can heavily affect projects done from the top-down, since core functions for the old specifications have already been planned/abstracted/coded.

b) What is a drawback of the bottom up approach?

One drawback of the bottom-up approach is that there is no working project until the end of the development cycle. This means it can be difficult to determine missing functions / what parts of the project have been finished. This also makes general testing of the project's functionalities difficult until all of the parts of the project are finished.

a) Complete the code below to read in a filepath to the `train.csv` and returns the DataFrame.

```python
import pandas as pd

df = pd.read_csv("train.csv")
df.describe()
```

b) Complete the code so it returns the number of rows that have at least one empty column value

```python
print("there are " +  str(df.isnull().any(axis=1).sum()) + " rows with at least one empty value")
```

c) Complete the code below to remove all columns with more than 200 NaN values

```python
df = df.dropna(axis=1, thresh=len(df.index)-200)
df.columns
```

d) Complete the code below to replaces `male` with 0 and `female` with 1

```python
df['Sex'] = df['Sex'].replace({'male': 0, 'female': 1})
df.head()
```

e) Complete the code below to add four columns `First Name`, `Middle Name`, `Last Name`, and `Title` corresponding to the value in the `name` column.

For example: `Braund, Mr. Owen Harris` would be: 

|First Name | Middle Name | Last Name | Title |
|-----------|-------------|-----------|-------|
| Owen      |  Harris     |  Braund   | Mr    |

Anything not clearly one of the above 4 categories can be ignored.

```python
df[['First Name', 'Middle Name', 'Last Name', 'Title']] = df['Name'].str.split(' ', n=3, expand=True)[[2, 3, 0, 1]].apply(lambda row: row.str.replace(',', '')) # last bit is just to remove the comma in last name
df.head()
```

f) Complete the code below to replace all missing ages with the average age

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df.head()
```
